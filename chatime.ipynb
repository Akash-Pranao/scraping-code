```python
headers = {
    'accept': '*/*',
    'accept-language': 'en-US,en;q=0.9',
    'priority': 'u=1, i',
    'referer': 'https://chatime.redcatcloud.com.au/app/',
    'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
}

response = requests.get(
    'https://chatime.redcatcloud.com.au/api/v1/stores/248/menu?&sale_type=102&ordering_provider=1',
    headers=headers,
)
menu_response = response.json()['data']

product_list = []
category_list = []

for category in menu_response['categories']:
    options = category.get('option_ids', [])
    for option in options:
        category_list.append({
            'item_category': category.get('name'),
            'item_id': option
        })

category_df = pd.DataFrame(category_list)

# Loop through each product in menu_response and extract details
for product in menu_response['products']:
    mods = product.get('products')
    mod_list = []
    if mods:
        for mod in mods:
            mod_list.append(mod.get('recid'))

    product_list.append({
         
        'item_id': product.get('recid'),
        'item_name': product.get('display_name'),
        'item_price': product.get('price', 0) / 100,
        'item_category_two': product.get('subcategory', None),
        'item_description': product.get('item_info'),
        'item_calories': product.get('kilojoules'),
        'size_modifiers' : mod_list
        # Adjust if needed
    })

# Convert the list to a pandas DataFrame
menu_df = pd.DataFrame(product_list)

exploded_menu = menu_df.explode('size_modifiers')

menu_df = menu_df[['item_id','item_name','item_price']]
final_menu_df = pd.merge(exploded_menu,category_df,on = ['item_id'])
final_menu_df = pd.merge(final_menu_df,menu_df,how = 'left',left_on = 'size_modifiers',right_on = 'item_id')
final_menu_df = final_menu_df.rename(columns={
    'item_id_x': 'item_id',
    'item_name_x': 'item_name',
    'item_name_y': 'size_options'
})

# Create the 'item_price' column based on the condition
final_menu_df['item_price'] = final_menu_df.apply(
    lambda row: row['item_price_x'] if row['item_price_x'] != 0 else row['item_price_y'], axis=1
)

# Drop unnecessary price columns if no longer needed
final_menu_df = final_menu_df.drop(columns=['item_price_x', 'item_price_y','item_id_y','size_modifiers'])

final_menu_df

    # 'item_id', 'item_name', 'item_price',
    #    'item_category', 'item_category_two', 'item_description',
    #    'item_calories',
```

###  MODIFIERS

```python
prod_opt = []
for product in menu_response['products']:
    choices = product.get('choicesets')
    if choices:
        for choice in choices:
            prod_opt.append({
                'item_id': product.get('recid'),
                'item_name': product.get('display_name'),
                'modifier_id': choice.get('recid')})
prod_opt_df = pd.DataFrame(prod_opt)
prod_opt_df
```

```python
# Initialize an empty list to store the modifier data
modifiers_list = []

# Loop through the 'choicesets' to extract modifier details
for option in menu_response['choicesets']:
    choices = option.get('choices')
    for choice in choices:
        modifiers_list.append({
            'modifier_id': choice.get('recid'),
            'modifier_category': option.get('display_name'),
            'modifier_price': choice.get('price',0)/100,
            'modifier_required': option.get('required'),
            'min_quantity': option.get('minimum'),
            'max_quantity': option.get('maximum'),
            'default_quantity': option.get('default_quantity', np.nan),
            'is_modifier_selected_by_default': choice.get('selected')
        })

# Convert the list of modifiers to a DataFrame
modifiers_df = pd.DataFrame(modifiers_list)

modifiers_df





# 'modifier_code', 'store_id', 'item_code', 'item_name', 'modifier_name',
    #    'modifier_price', 'modifier_group', 'modifier_calories',
    #    'modifier_required', 'min_quantity', 'max_quantity', 'default_quantity',
    #    'is_modifier_selected_by_default'
```

```python
final_modifier_df = pd.merge(prod_opt_df,modifiers_df,on = 'modifier_id',how ='left')
final_modifier_df
```
